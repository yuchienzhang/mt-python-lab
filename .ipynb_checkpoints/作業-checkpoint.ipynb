# 分析國家GDP與自殺率之關係

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

import matplotlib as mpl
mpl.rc('font', family='DejaVu Sans')
```

```python
df = pd.read_csv('/Users/yuchienchang/Downloads/master.csv')
```

## 找出2014年男性青少年自殺資料

```python
df = df[(df.year == 2014) & (df.sex == 'male') & (df.age == '15-24 years')]
```

```python
df
```

## 重設index

```python
df.index = range(1,79)
```

```python
df
```

## 進行回歸分析

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
x = np.array(df['gdp_per_capita ($)'])
y = np.array(df['suicides/100k pop'])
x = x.reshape(len(x), 1)
plt.scatter(x, y)
```

### 七成資料當作訓練資料

```python
regr = LinearRegression()
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 1029)
regr.fit(x_train, y_train)
y_pred = regr.predict(x_test)
```

```python
plt.scatter(x_test, y_test)
plt.plot(x_test, y_pred, 'r')
```

```python
regr.score(x_test, y_test)
```
