```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## 設計一個loss function

```python
x = np.linspace(-1, 1, 100)
def fn(x):
    return -(20*x**4 + 30*x**3 - 100*x**2 + x +6)
y = fn(x)
plt.plot(x, y)
```

## Gradient Descent

```python
def gradient_dscent(x_start, epochs, learning_rate):  
    run = []
    y_run = []
    x = x_start
    run.append(x)
    y_run.append(fn(x))
    for i in range(epochs):
        dx = -80*x**3 - 90*x**2 + 200*x -1 ##自己手動微分
        x = x - learning_rate * dx  ##往斜率的反方向走
        run.append(x)
        y_run.append(fn(x))
        plt.scatter(run[i], y_run[i], c='r')
```

```python
ham = gradient_dscent(0.75, 50, 0.001) ##自訂起始點、迭代次數、學習率
plt.plot(x, y)
```
