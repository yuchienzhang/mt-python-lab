```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## 唬爛一個loss function

```python
x = np.linspace(-1, 1, 100)
y = -(20*x**4 + 30*x**3 - 100*x**2 + x +6)
plt.plot(x, y)
```

## Gradient Descent

```python
def gradient_dscent(x_start, epochs, learning_rate):  
    used = []    
    x = x_start   
    used.append(x)
    for i in range(epochs):
        dx = -80*x**3 - 90*x**2 + 200*x -1 ##自己手動微分
        x = x - learning_rate * dx  
        used.append(x)
    return used[-1]
```

```python
ham = gradient_dscent(0.25, 50, 0.001)
print(ham)
```

```python
plt.plot(x, y)   
plt.scatter(ham, -(20*ham**4 + 30*ham**3 - 100*ham**2 + ham +6), c='r')
```
