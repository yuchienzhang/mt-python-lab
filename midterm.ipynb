## 現在我得到了各科系學士班畢業的平均薪資，以及102年度全台大專院校就讀該科系的人數。接著我想用回歸分析觀察兩變數之間的關係。

```python
import matplotlib.pyplot as plt
```

```python
 for i in range(len(dfnew)):
        dfnew['學士日間學制平均薪資'][i] = dfnew['學士日間學制平均薪資'][i].replace(',', '')
```

```python
x = []
y = []
for i in range(len(dfnew)):
    x.append(int(dfnew['學士日間學制平均薪資'][i]))
    y.append(int(dfnew['學生總數'][i]))
    
plt.scatter(x, y)
```

## 我天真的以為畢業薪資跟就讀人數會呈正相關，也就是說大家看到某個科系畢業後的薪水待遇很好，會爭先恐後地去就讀那個科系，結果畫出來的圖似乎也不用分析了，用肉眼看就覺得沒什麼關係。

### 看來有個科系在很右邊，平均薪資超過九萬但是就讀人數僅一萬多人，看一下是什麼系好了

```python
for i in range(len(dfnew)):
    if int(dfnew['學士日間學制平均薪資'][i]) > 90000:
        print(dfnew.loc[i])
```
